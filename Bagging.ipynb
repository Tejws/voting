```python
# Step 1: Import required libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_breast_cancer  # Changed to load_breast_cancer instead of load_wine
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
from sklearn import datasets

# Step 2: Load the Breast Cancer dataset
cancer = datasets.load_breast_cancer()  # Load Breast Cancer dataset instead of Wine
X = cancer.data
y = cancer.target

# Step 3: Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 4: Create the base classifier (Decision Tree Classifier)
base_clf = DecisionTreeClassifier(random_state=42)

# Step 5: Create the BaggingClassifier (no base_estimator argument needed anymore)
bagging_clf = BaggingClassifier(base_clf, n_estimators=50, random_state=42)

# Step 6: Train the model using the training data
bagging_clf.fit(X_train, y_train)

# Step 7: Make predictions on the test set
y_pred = bagging_clf.predict(X_test)

# Step 8: Evaluate the model performance
accuracy = accuracy_score(y_test, y_pred)

# Print the accuracy
print(f'Accuracy of Bagging Classifier: {accuracy * 100:.2f}%')

# Step 9: Optionally, plot a confusion matrix to better understand the performance
from sklearn.metrics import confusion_matrix
import seaborn as sns

cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=cancer.target_names, yticklabels=cancer.target_names)
plt.title('Confusion Matrix for Bagging Classifier')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```
